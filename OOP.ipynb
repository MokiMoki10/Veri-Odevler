```python
#Öğrenci Sınıfı
class Ogrenci:
    def __init__(self,isim,no,grade):
        if min(grade)<0 or max(grade)>100:
            raise ValueError("Girdiğiniz değerler 0-100 arasında olmalı")
        self.isim = isim
        self.no = no
        self.grade = grade
        self.ort = sum(grade)/len(grade)
    def __str__(self):
        return(f"Öğrenci:{self.isim} Numara:{self.no} Notlar:{self.grade} Ortalama:{self.ort}")  
a = Ogrenci("Ali",12345,[85,90,95])
print(a)
```

```python
#Kitap Sınıfı
class Book:
    def __init__(self, isim, yazar, sayfa):
        if sayfa <= 0:
            raise ValueError("Sayfa sayısı negatif olamaz")
        self.isim = isim
        self.yazar = yazar
        self.sayfa = sayfa
        self.okunan_sayfa = 0 
    def oku(self, artis):        
        if self.okunan_sayfa + artis > self.sayfa:
            raise ValueError("Sayfa sayısından fazlasını okuyamazsın")
        self.okunan_sayfa = self.okunan_sayfa + artis
    def __str__(self):
        return(f"{self.isim} by {self.yazar} - {self.sayfa} sayfa, {self.okunan_sayfa} sayfa okundu")
kitap1 = Book("ABC", "SSSSS", 300)
print(kitap1)
kitap1.oku(100) 
print(kitap1)
```

```python
#Banka Sınıfı
class Bank:
    def __init__(self,no,ad,bakiye,yatir,cek):
        self.no = no
        self.ad = ad
        self.bakiye = bakiye + yatir - cek
        self.yatir = yatir
        self.cek = cek
        self.sonislem = yatir-cek
    def parayatir(self,x):
        if x<0:
            raise ValueError ("Negatif olamaz")
        self.bakiye = self.bakiye + x
        self.sonislem = +x
        print(f"Bakiye: {self.bakiye}TL, Son yapılan işlem:+{x}") 
    def paracek(self,y):
        if y < 0:
            raise ValueError ("Negatif olamaz")
        if y > self.bakiye:
            raise ValueError ("O kadar paran yok")
        self.bakiye = self.bakiye - y
        self.sonislem = -y
        print(f"Bakiye: {self.bakiye}TL, Son yapılan işlem:-{y}") 
    def __str__(self):
        return(f"İsim: {self.ad}, Hesap NO: {self.no}, Bakiye: {self.bakiye}TL, Son işlem: {self.sonislem}")
a1 = Bank(12345,'Ahmet Yalçın', 1200, 500, 300)
print(a1)
a1.parayatir(300)
print(a1)
a1.paracek(500)
print(a1)
```
