```python
#50 adet noktanın 4 boyutlu uzaydaki koordinatlarını içeren (50, 4) boyutunda bir matris (A) ve 1 adet referans noktasını içeren (1, 4) boyutunda bir matris (B) oluşturun.
a = np.random.randn(50,4)
b = np.random.randn(1,4)
cikart = a - b 
kare_al = cikart ** 2
topla = np.sum(kare_al, axis = 1)
kokten_cikar = np.sqrt(topla)
print(kokten_cikar)
yakin = np.min(kokten_cikar)
print(f"En yakın noktanın mesafesi: {yakin}")
indeks = np.argmin(kokten_cikar)
print(f"A makrisindeki indeksi {indeks}")
```

```python
#Finansal Volatilite Matrisi (Lineer Cebir & Dif)
matris = np.random.uniform(80,200,720).reshape(120,6)
fark = np.diff(matris,axis = 0)
degisim = fark / matris[:-1]
print(f"Yüzde değişim {degisim[:3]}")
cov = np.cov(degisim,rowvar = False)
print(f"Koveryans matrisi:\n {cov}") 
value,vector = np.linalg.eig(cov)
indeks = np.argmin(value)
en_kucuk_val = value[indeks]
en_kucuk_vek = vector[:,indeks]
print(en_kucuk_val)
print(en_kucuk_vek)
```

```python
#İklim Anomali Tespiti (3D Eksen Yönetimi)
matris = np.random.randint(15,35,size=(10,30,2))
matris[:,:,1] = np.random.randint(50,80,size=(10,30))
ort = np.mean(matris,axis=(0,1))
print("Ortalama sıcaklık", ort[0])
print("Ortalama nem", ort[1])
vector = ort.reshape(1,1,2)
print("Ortalama vektörü",vector)
yeni_matris = matris - vector
print(yeni_matris)
```

```python
#NaN Değerlerini Koşullu Doldurma (Missing Data Imputation)
matris = np.random.randn(20, 5)
rastgele_satirlar = np.random.choice(20, size=10, replace=True)
rastgele_sutunlar = np.random.choice(5, size=10, replace=True)

for i in range(10):
    matris[rastgele_satirlar[i], rastgele_sutunlar[i]] = np.nan
medyan = np.nanmedian(matris, axis=0)
print("Sütunların medyanları\n",medyan)
yeni = np.where(np.isnan(matris),medyan,matris)
print(yeni)
print(f"\nDoldurulmuş matristeki NaN sayısı: {np.sum(np.isnan(yeni))}")

```

```python
#Özel Birleştirme ve İndeksleme
a = np.random.randint(0,100,size=(10,3))
b = np.random.randint(0,100,size=(10,3))
yeni = np.c_[a[:,0],b[:,1],a[:,2]]
print(yeni)
medyan = np.median(yeni[:,0])
print("Medyan:",medyan)
sapma = np.std(yeni[:,1])
print("Standard sapma: ",sapma)
toplam =np.sum(yeni[:,2])
print("Toplam:",toplam)
vektör = np.array([medyan,sapma,toplam])
print(vektör)
```
