```python
#1. Listedeki En Büyük Sayıyı Bulma
def find_max(n):
    if len(n) == 0:
        return ("Listeniz boş olamaz")
    else:
        toplam = 0
        for i in range(0,len(n)):
            toplam=toplam+n[i]
        return toplam    
n=[1,3,-5,18,3,-10,5]
print(find_max(n))
```

```python
#2. String'i Ters Çevirme
def ters_cevir(n):
    if type(n) != str: 
        return "Girdiğiniz değer string olmalıdır"
    else:
        return n[::-1]
n="Python"
print(ters_cevir(n))
```

```python
#3. Faktöriyel Hesaplama
def factorial(n):
    if n < 0:
        return ("Girdiğin değer 0 veya pozitif olmalı")
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(-5))        
```

```python
#4. Listedeki Çift Sayıları Filtreleme
def cift_arama(n):
    if len(n) == 0:
        return "Girdiğiniz liste boş olamaz"
    else:
        sonuc=[]
        for i in n:
            if i % 2 == 0:
                sonuc.append(i)
        return sonuc        
n=[1,3,4,6,7,10]
print(cift_arama(n))
```

```python
#5. İki Listenin Birleşimi ve Tekilleştirilmesi
def birlestir(n,m):
    if len(n) == 0 and len(m) == 0:
        return "İki liste de boş olamaz"
    else:
        liste=[]
        for i in n:
            if i in liste: 
                continue
            else:
                liste.append(i)
        for i in m:
            if i in liste:
                continue
            else:
                liste.append(i)
        return liste
n=[1,3,5,6,7]
m=[10,3,77,3,9,7]
print(birlestir(n,m))
```

```python
#6. Palindrom Kontrolü
def pal_check(n):
    s = "".join(n.split()).lower()        
    return s == s[::-1]
print(pal_check("A man a plan a canal Panama")) 
```

```python
#7. İki Sayının En Büyük Ortak Böleni (EBOB)
def ebob(a,b):
    if a < 0 or b < 0:
        return "Girdiğin değerler pozitif olmalı"
    else:
        while b>0:
            a,b=b,a%b
        return a
print(ebob(60,200))        
        
```

```python
#8. Anagram Kontrolü
def anagram(n,m):
    return sorted(n) == sorted(m)
n="listen "
m="silent"
print(anagram(n,m))
```

```python
#9. Listedeki Eksik Sayıyı Bulma
def eksik_bul(n):
    if max(n) - len(n) != 1:
        return "Liste hatalı"
    else:
        for i in range (1,max(n)):
            if i not in n:
                return i
n = [1, 3, 4, 5,]
print(eksik_bul(n))
```

```python
#10. Listeyi Belirli Bir Değere Göre Döndürme (Rotate)
def rotate(n,k):
    if k < 0: 
        return "Girdiğin k değeri negatif olamaz"
    else:
        k = k % len(n)
        return n[-k:] + n[:-k]
n = [1,3,5,8,10,16]    
print(rotate(n,3))
    
```

```python
#12. En Uzun Palindromik Alt String
def pal_finder(n):
    if len(n)>1000 and " " in n:
        return "Girdiğin string uygun değil"
    else:
        maks=""
        for i in range(0,len(n)):
            for j in range (i+1,len(n)):
                s = n[i:j]
                if s == s[::-1] and len(maks) < len(s):
                    maks = s
        return maks 
print(pal_finder('efebabdefe'))
```

```python
#14. Maksimum Alt Dizi Toplamı
def max_subset(n):
    if len(n) == 0: 
        return "Liste boş olamaz"
    else:
        maks = sum(n)
        for i in range (0,len(n)):
            for j in range (i+1,len(n)):
                if maks < sum(n[i:j]): 
                    maks = sum(n[i:j])
    return maks 
n = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(max_subset(n))
```
